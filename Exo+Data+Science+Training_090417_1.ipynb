```python
def match_ends(words):
    a,b = 0,0
    for word in words:
        if len(word) >= 2:
            a += 1
        if word[0] == word[-1]:
            b += 1
    
    print('the number of strings where the string length is 2 is', a)
    print('the number of strings where the first and last chars of the string are the same is ', b)
    return
```

```python
test=['zxer', 'xerex', 'rtuy']
```

```python
match_ends(test)
```

```python
def front_x(words):
    liststd = []
    listx = []
    for word in words:
        if word[0] == 'x':
           listx.append(word)
        else:
            liststd.append(word)
    
    listx.sort()
    liststd.sort()
    print(listx + liststd)
    return
```

```python
test=['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] 
```

```python
front_x(test)
```
