```python
def match_ends(words):
    a,b = 0,0
    for word in words:
        if len(word) >= 2:
            if word[0] == word[-1]:
                b += 1
    return b
```

```python
test=['zxer', 'xerex', 'rtuy']
```

```python
match_ends(test)
```

```python
def front_x(words):
    liststd = []
    listx = []
    for word in words:
        if word[0] == 'x':
           listx.append(word)
        else:
            liststd.append(word)
    
    listx.sort()
    liststd.sort()
    print(listx + liststd)
    return
```

```python
test=['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] 
```

```python
front_x(test)
```

```python
def sort_last(tuples):
    return sorted(tuples, key=lambda x:x[-1])
```

```python
test=[(1,7),(1,3),(2,2)]
```

```python
sort_last(test)
```

```python
def remove_adjacent(nums):
    nums_dedup=[]
    for index in range(len(nums)):
        if nums[index] != nums[-1]:
            if nums[index] != nums[index+1]:
                nums_dedup.append(nums[index])
        else:
                nums_dedup.append(nums[index])
    return nums_dedup
```

```python
test=[1, 2, 2, 3, 3, 4]
```

```python
remove_adjacent(test)
```

```python
def linear_merge(list1, list2):
    return sorted((sorted(list1, key=lambda x:x[0]) + sorted(list2, key=lambda x:x[0])), key=lambda x:x[0])
```

```python
list1=['aa','aa','bb']
```

```python
list2=['zz', 'xx', 'bb']
```

```python
linear_merge(list1, list2)
```

```python
def donuts(count):
    if count >= 10:
        a = "Number of donuts: many"
    else:
        a = "Number of donuts: " + str(count)
    return a
```

```python
donuts(5)
```

```python
def both_ends(s):
    return (s[:2]) + (s[-2:])
```

```python
def fix_start(s):
    return s[0] + s[1:].replace(s[0],"*")
```

```python
fix_start('babble')
```

```python
def mix_up(a,b):
    return (b[:-1] + a[-1:]) + ' ' + (a[:-1] + b[-1:])
```

```python
mix_up('mix','pod')
```

```python
def verbing(s):
    if len(s) > 3:
        if s[-3:] == "ing":
            s += "ly"
        else:
            s += "ing"
    return s
```

```python
verbing ('speak')
```

```python
verbing('speaking')
```

```python
verbing('pee')
```

```python
def not_bad(s):
    car1="not"
    car2="bad"
    if car1 in s:
        if car2 in s[(s.find(car1)):]:
            return s[:(s.find(car1))] + 'good'
```

```python
not_bad('This dinner is not that bad')
```
