```python
def match_ends(words):
    a,b = 0,0
    for word in words:
        if len(word) >= 2:
            if word[0] == word[-1]:
                b += 1
    return b
```

```python
test=['zxer', 'xerex', 'rtuy']
```

```python
match_ends(test)
```

```python
def front_x(words):
    liststd = []
    listx = []
    for word in words:
        if word[0] == 'x':
           listx.append(word)
        else:
            liststd.append(word)
    
    listx.sort()
    liststd.sort()
    print(listx + liststd)
    return
```

```python
test=['mix', 'xyz', 'apple', 'xanadu', 'aardvark'] 
```

```python
front_x(test)
```

```python
def sort_last(tuples):
    return sorted(tuples, key=lambda x:x[-1])
```

```python
test=[(1,7),(1,3),(2,2)]
```

```python
sort_last(test)
```

```python
def remove_adjacent(nums):
    nums_dedup=[]
    for index in range(len(nums)):
        if nums[index] != nums[-1]:
            if nums[index] != nums[index+1]:
                nums_dedup.append(nums[index])
        else:
                nums_dedup.append(nums[index])
    return nums_dedup
```

```python
test=[1, 2, 2, 3, 3, 4]
```

```python
remove_adjacent(test)
```

```python
def linear_merge(list1, list2):
    return sorted((sorted(list1, key=lambda x:x[0]) + sorted(list2, key=lambda x:x[0])), key=lambda x:x[0])
```

```python
list1=['aa','aa','bb']
```

```python
list2=['zz', 'xx', 'bb']
```

```python
linear_merge(list1, list2)
```
